````python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
````

````python
documents = [
    Document(page_content="LangChain helps developers build LLM applications easily."),
    Document(page_content="Chroma is a vector database optimized for LLM-based search."),
    Document(page_content="Embeddings convert text into high-dimensional vectors."),
    Document(page_content="OpenAI provides powerful embedding models."),
]
````

# ***MMR***

###⚖️ **MMR Retriever** (Maximal Marginal Relevance)

- **Purpose**: Diversify retrieved documents to avoid redundancy. How can we pick results for the query that are not only relevant but also different from each other.

- **Method**: Balances relevance and dissimilarity
- **Use Case**: When results are too similar or redundant

- picks the most relevant document first --> then picks the most relevant and least simillar to already selected docs.

````python
# Sample documents
docs = [
    Document(page_content="LangChain makes it easy to work with LLMs."),
    Document(page_content="LangChain is used to build LLM based applications."),
    Document(page_content="Chroma is used to store and search document embeddings."),
    Document(page_content="Embeddings are vector representations of text."),
    Document(page_content="MMR helps you get diverse results when doing similarity search."),
    Document(page_content="LangChain supports Chroma, FAISS, Pinecone, and more."),
]
````

````python
from langchain.vectorstores import FAISS

embedding_model = hf

vectorstore_faiss = FAISS.from_documents(docs, embedding=hf)
````

````python
retriever = vectorstore_faiss.as_retriever(search_type="mmr", search_kwargs ={"k":3,"lambda_mult":0.5})
````

````python
query = "what is langchain"
````

````python
retriever.invoke(query)
````

## MULTI QUERY RETRIEVER

 🔁 Multi-Query Retriever in LangChain

## 🧠 Motivation

When you ask an LLM-based system a question like:

> "How does LangChain support vector stores?"

You might miss relevant content if:
- The phrasing is narrow
- Keywords are absent in documents
- Synonyms or paraphrases are used

🔎 **MultiQueryRetriever** solves this by generating **multiple query variations** — improving recall and result diversity.

---

## 🔍 What is MultiQueryRetriever?

> A retriever that uses an LLM to generate **diverse rephrasings** of a query and then performs retrieval for each of them.

- 🚀 Increases **recall** by covering a wider semantic space
- 🧠 Uses an LLM to generate alternate phrasings
- 🔗 Uses a base retriever (like FAISS or BM25) for the actual search

---


User Query → LLM → Multiple Reformulated Queries → Retriever → Merged Results



For example:
> Original: `"LangChain and vector databases"`

Generated variations:
- `"What vector databases work with LangChain?"`
- `"Does LangChain support FAISS or Pinecone?"`
- `"Compatible vector stores with LangChain"`

Each is sent to the base retriever, and results are combined.

---

## 🧪 Example Code

```python
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chat_models import ChatOpenAI

# Embedding and vectorstore setup
embeddings = HuggingFaceEmbeddings()
vectorstore = FAISS.load_local("index_path", embeddings)

# Base retriever
base_retriever = vectorstore.as_retriever()

# LLM for generating query variations
llm = ChatOpenAI(model="gpt-3.5-turbo")

# Multi-query retriever setup
multi_retriever = MultiQueryRetriever.from_llm(
    retriever=base_retriever,
    llm=llm
)

# Run retrieval
docs = multi_retriever.get_relevant_documents("How does LangChain support vector search?")
for doc in docs:
    print(doc.page_content)

## ContextualCompressionRetriever

# 🧹 Contextual Compression Retriever in LangChain

## 🧠 Why Contextual Compression?

When retrieving documents in RAG, sometimes the results are:
- Too long 📜
- Contain irrelevant information ❌
- Have only **small relevant chunks** hidden inside

👉 This wastes **token budget** and reduces **response quality**.

---

## 🧰 What is ContextualCompressionRetriever?

> A wrapper retriever that **compresses** the retrieved documents using a **filter or summarizer**, typically an LLM or heuristic.

🔧 It consists of:
- A **base retriever** (e.g., FAISS, BM25)
- A **compressor** (e.g., LLMChain, Cohere, or custom filter)

📦 The result? You only pass **highly relevant, compressed context** to the LLM.

---


## 🔄 How It Works

Query → Base Retriever → Retrieved Docs → Compressor (LLM or Filter) → Final Slimmed-Down Docs

---

## ⚙️ Example Code: Using LLM for Compression

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# Embedding and retriever setup
embeddings = HuggingFaceEmbeddings()
vectorstore = FAISS.load_local("index_path", embeddings)
base_retriever = vectorstore.as_retriever()

# Compressor using LLMChain
llm = ChatOpenAI(model="gpt-4")

# Prompt for extracting relevant content
prompt = PromptTemplate.from_template(
    "Extract only the parts relevant to answering: {query}\n\n{text}"
)

llm_chain = LLMChain(llm=llm, prompt=prompt)
compressor = LLMChainExtractor.from_llm_chain(llm_chain)

# ContextualCompressionRetriever setup
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor,
    base_retriever=base_retriever
)

# Run it
docs = compression_retriever.get_relevant_documents("How does LangChain integrate with FAISS?")
for doc in docs:
    print(doc.page_content)


🔁 Other Compressor Types


| Compressor                   | Description                              |
| ---------------------------- | ---------------------------------------- |
| `LLMChainExtractor`          | Extracts only relevant text using LLM    |
| `CohereRerank`               | Reranks and trims with Cohere model      |
| `DocumentCompressorPipeline` | Chains multiple compressors together     |
| `LLMReducer`                 | Summarizes the full document             |
| `KeywordFilter`              | Keeps only chunks with specific keywords |


### 🧠 Summary
- ContextualCompressionRetriever refines raw retrieved docs
- Uses a compressor (LLM or heuristic) to trim noise
- Improves response quality and context relevance
